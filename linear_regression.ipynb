```python
# libraries import kar rahe hain
import pandas as pd
import seaborn as sns
```

```python
# data load karo
df = pd.read_csv("insurance.csv")
```

```python
# data dekho
df
```

```python
# bmi aur charges ka graph banana
sns.scatterplot(x=df["bmi"], y=df["charges"], hue=df["smoker"])
```

```python
# features aur target alag karo
features = df.drop(columns=["charges", "region"])
target = df["charges"]

# text columns ko numbers mein badlo
features["sex"] = features["sex"].map({"female": 1, "male": 0})
features["smoker"] = features["smoker"].map({"yes": 1, "no": 0})
```

```python
# pehle 5 rows dekho
features.head()
```

```python
# data ko train aur test mein baanto
from sklearn.model_selection import train_test_split

train_size = 0.8
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=1-train_size, random_state=42
)
```

```python
# test data dekho
X_test.head()
```

```python
# Linear Regression model banao aur train karo
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# test data pe predict karo
predictions = reg.predict(X_test)
print(predictions)
```

```python
# actual values dekho
y_test
```

```python
# model ka score check karo
from sklearn.metrics import r2_score

score = r2_score(y_test, predictions)
print("R2 Score:", score)

# adjusted r2 nikalo
total_samples = X_test.shape[0]
num_features = X_test.shape[1]

adj_score = 1 - ((1 - score) * (total_samples - 1) / (total_samples - num_features - 1))
print("Adjusted R2 Score:", adj_score)
```
